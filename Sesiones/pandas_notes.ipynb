# Creación de DataFrame (tablas)
Para este ejemplo sencillo vamos a recrear la siguiente tabla:

| Station Name   | Code |  PM10 |   SR  |
|----------------|------|:-----:|:-----:|
| Obispado       | CE   | 94.84 | 99.09 |
| Escobedo       | N    | 94.11 | 97.58 |
| Universidad    | N 2  |  94.3 | 98.55 |
| San Nicolas    | NE   | 97.37 | 99.76 |
| Apodaca        | NE 2 | 96.08 | 99.39 |
| García         | NW   | 94.86 | 99.49 |
| San Bernabé    | NW 2 | 96.64 | 99.49 |
| Santa Catarina | SW   | 92.61 | 99.21 |
| San Pedro      | SW 2 | 95.09 |  99.5 |

En donde primero introduciremos el nombre de las columnas, y los indices, con lo cual el resultado esperado es el siguiente:

| Station Name   | Code |  PM10 |   SR  |
|----------------|------|:-----:|:-----:|
| Obispado       | NaN  |  Nan  |  NaN  |
| Escobedo       | NaN  |  Nan  |  NaN  |
| Universidad    | NaN  |  Nan  |  NaN  |
| San Nicolas    | NaN  |  Nan  |  NaN  |
| Apodaca        | NaN  |  Nan  |  NaN  |
| García         | NaN  |  Nan  |  NaN  |
| San Bernabé    | NaN  |  Nan  |  NaN  |
| Santa Catarina | NaN  |  Nan  |  NaN  |
| San Pedro      | NaN  |  Nan  |  NaN  |

````python
"""
Definimos en una lista el nombre de los indices y columnas de la tabla
"""
columns=[
         "Code",
         "PM10",
         "SR"]
index=["Obispado",
       "Escobedo",
       "Universidad",
       "San Nicolas",
       "Apodaca",
       "Garcia",
       "San Bernabe",
       "Santa Catarina",
       "San Pedro"]
data=pd.DataFrame(index=index,columns=columns)
data
````

````python
"""
Ahora pondremos en una lista los valores de cada columna
"""
stations_code=[
    "CE",
    "N",
    "N 2",
    "NE",
    "NE 2",
    "NW",
    "NW 2",
    "SW",
    "SW 2"]
pm10_data=[
    94.84,
    94.11,
    94.3,
    97.37,
    96.08,
    94.86,
    96.64,
    92.61,
    95.09,]
sr_data=[
    99.09,
    97.58,
    98.55,
    99.76,
    99.39,
    99.49,
    99.49,
    99.21,
    99.5,]
````

````python
"""
Introduciremos estos valores dentro del dataframe
"""
data["Code"]=stations_code
data["PM10"]=pm10_data
data["SR"]=sr_data
data
````
