## 4.1 Estructura If


Se usa para tomar una decisión en caso de que una condición sea verdadera,
podemos visualizarlo de manera de la siguiente manera:



If.png

Toma la siguiente estructura:



```
# if (condición):
#   Acción
# Siguiente linea de código
```

If se encarga verificar que una condición arbitraria sea verdadera, de serlo, ejecuta una instrucción brindada por el usuario. Si resulta ser que la condición no se cumple entonces se pasa a la siguiente linea de código.


Veamoslo en ejemplos:


````python
# EJEMPLO 1
# MAYOR QUE

num = 3

if num > 0:
    print("{} es un número positivo".format(num))

print("Este mensaje siempre será impreso")
````

````python
# EJEMPLO 2
# MENOR QUE

a = 33
b = 200
if b > a:
  print("b es mayor que a")


c=12
d=-23
if d > c:
  print("d es mayor que c")
````

4.2 Estructura If-else


Se usa para tomar una decisión A en caso de que una condición sea verdadera y una decisión B en caso de que sea falsa la condición, podemos visualizarlo de manera de la siguiente manera:

If_Else.png

Toma la siguiente estructura:



```
# if (condición):
#   Acción cuando la condición es verdadera
# else:
#   Acción cuando la condición es falsa
# Siguiente linea de código
```

If se encarga verificar que una condición arbitraria sea verdadera, de serlo, ejecuta una instrucción brindada por el usuario.

Si resulta ser que la condición no se cumple entonces se pasa a la siguiente linea de código.


Veamos en ejemplos:


````python
# EJEMPLO 3
# Números negativos

num = -5

if num >= 0:
    print("No es un número negativo")
else:
    print("Es un número negativo")
````

````python
# EJEMPLO 4
# Números pares

num = 7

if (num%2 == 1):
    print("No es un número par")
else:
    print("Es un número par")
````

## 4.3 Elif


Nos permite ahorrar lineas de código y evitar una excesiva identación en el código, funciona de la siguiente manera:

Tenemos el siguiente segmento de código

```
# if (condición_1):
#     (instrucción_1)
# else: 
#     if (condición_2):
#         (instrucción_2)
#     else: 
#         (instrucción_3) 
```

Podemos acortarlo de la siguiente manera

```
# if (condición_1):
#     (instrucción_1)
# elif (condición_2):
#     (instrucción_2)
# else: 
#     (instrucción_3)
```

Ambos representan la siguiente estructura:

Elif.jpg


````python
# EJEMPLO 6
# CLASIFICACIÓN NEGATIVOS, POSITIVOS Y EL CERO

num = 0


if num < 0:
    print("Número negativo")
elif num != 0:
    print("Número positivo")
else:
    print("El número es cero")
````
