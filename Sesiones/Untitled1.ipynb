```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# function that returns dy/dt
def model(y,x):
    dy=(2*x*y[1]-(y[1])**3)/(2*x)
    return (y[1],dy)

# initial condition
y0 = [5,2]

# time points
t = np.linspace(1,11,100)

# solve ODE
solution = odeint(model,y0,t)

# plot results
plt.plot(t,solution)
plt.xlabel('time')
plt.ylabel('y(t)')
plt.show()
```
