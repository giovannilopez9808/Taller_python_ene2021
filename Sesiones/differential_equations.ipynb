### Ecuaciones diferenciales

##### Ecuación diferencial de primer orden

De forma general, una ecuación de primer orden puede ser escrita de la siguiente manera:
\begin{equation*}
y'+yP(x,y)=Q(x,y)
\end{equation*}

Por ejemplo, se tiene el siguiente problema:

Resuelva la ecuación diferencial

\begin{equation*}
\frac{dy}{dx}+y=0
\end{equation*}

para todo $x>0$ con la condicion inicial $y(0)=2$. 

Para resolver este problema de manera análitica, el proceso seria el siguiente:
\begin{align*}
\frac{dy}{dx}+y&=0\\
\frac{dy}{dx}&=-y\\
\frac{dy}{y}&=-dx\\
ln(y)&=-x+c\\
y&=e^{-x+c}\\
y&=e^{-x}e^{c}\\
y&=Ce^{-x}
\end{align*}

Por lo tanto, la familia de soluciones es:
\begin{equation*}
y=e^{-x}
\end{equation*}


Aplicando la condición inicial, se obtiene que:
\begin{align*}
2&=Ce^{-0}\\
2&=C
\end{align*}

Por lo tanto, la solución particular es
\begin{equation*}
y=2e^{-x}
\end{equation*}


Para resolver esto con python, usaremos la libreria ``scipy``, en especifico su modulo ``integrate``, para usarlo tenemos que importarlo de la siguiente manera

````
from scipy.integrate import *
````

y para resolver ecuaciones diferenciales usaremos el siguiente comando:
````
solution = odeint(df,y0,x)
````
donde:
df es la ecuación diferencial de la forma
\begin{equation*}
\frac{dy}{dx}=f(x,y)
\end{equation*}
y0 son las condiciones iniciales y x son los valores donde queremos obtener las soluciones.

`````python
import matplotlib.pyplot as plt
from scipy.integrate import *
import numpy as np


# Ecuación diferencial de primer orden
def df(y, x):
    return -y


# Solución exacta obtenida en la celda de arriba
def f_sol(x):
    return 2 * np.exp(-x)


# Condicion inicial
y0 = 2
# Valores de la función
x = np.linspace(0, 2, 100)
# Modelo para resolver la ecuación diferencial
solution_num = odeint(df, y0, x)
# Solucion exacta
solution_ana = f_sol(x)
# plot results
plt.plot(x, solution_num, label="Numérica")
plt.plot(x, solution_ana, label="Análitica")
plt.xlabel('time')
plt.ylabel('y(t)')
plt.legend(frameon=False, loc="upper right")
plt.text(1.25,1.25,"$\\frac{dy}{dx}+y=0$",fontsize=16)
`````

##### Ecuaciones diferenciales de segundo orden
Una ecuación diferencial lineal de segundo orden es aquella que se puede escribir en la forma:
\begin{equation*}
\frac{d^2 y}{dx^2}+P(x)\frac{dy}{dx}+q(x)y=g(x)
\end{equation*}

Por ejemplo, se tiene la siguiente ecuación diferencial:
\begin{equation*}
\frac{d^2 y}{dx^2}-5\frac{dy}{dx}+6y=0
\end{equation*}

Para este caso, la solución general es el siguiente:
\begin{equation*}
y=Ae^{3x}+Be^{2x}
\end{equation*}

Si se tienen las condiciones iniciales y(0)=3 y y'(0)=2, entonces la solución particular es 
\begin{equation*}
y=-4e^{3x}+7e^{2x}
\end{equation*}

Para resolver esta ecuación diferencial con python, se realiza el siguiente procedimiento:
2xy''+(y')^3=2xy'

`````python
import matplotlib.pyplot as plt
from scipy.integrate import *
import numpy as np


# Ecuación diferencial a resolver
def f(y, x):
    dy = y[1]
    ddy = 5 * dy - 6 * y[0]
    return (dy, ddy)


# Solucion exacta de la ecuación diferencial
def f_sol(x):
    return -4 * np.exp(3 * x) + 7 * np.exp(2 * x)


# Condiciones iniciales
y0 = [3, 2]
# Valores para calcular la solución
x = np.linspace(0, 1, 20)
# Solucion de la ecuacion diferencial
solution_num = odeint(f, y0, x)
solution_ana = f_sol(x)

# plot results
plt.plot(x, solution_num[:, 0], label="Numérica")
plt.plot(x, solution_ana, label="Análitica")
plt.xlabel('x')
plt.ylabel('y(x)')
plt.legend(frameon=False, loc="upper right")
plt.text(0.1, -20, "$\\frac{d^2 y}{dx^2}-5\\frac{dy}{dx}+6y=0$",fontsize=16)
`````
